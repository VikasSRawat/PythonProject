```python
import numpy as np
# import pandas as pd
import matplotlib.pyplot as plt
# import matplotlib.axes as ax
import matplotlib
import pylab
from sklearn import svm
from sklearn.datasets import make_blobs
# create 40 seperable points
x,y=make_blobs(n_samples=40,centers=2,random_state=20) # This will generate a set(40 lines) of data which will be divided into 2 blobs randomly
# fitting the model
clf=svm.SVC(kernel='linear',C=1)
clf.fit(x,y)
plt.scatter(x[:,0],x[:,1],c=y,s=30,cmap=plt.cm.Paired)
plt.show()
```

```python
# Predicting where new data will fit inside the SVM
newdata=[[3,4],[5,6]]
print(clf.predict(newdata)) # Note 0 - blue  ,  1 - brown
```

```python
clf=svm.SVC(kernel='linear',C=1000)
clf.fit(x,y)
plt.scatter(x[:,0],x[:,1],c=y,s=30,cmap=plt.cm.Paired)

ax=plt.gca()
# get_xlim and get_ylim will get the coordinates of the starting and ending limit of the axes present in ax
xlim=ax.get_xlim()
ylim=ax.get_ylim()
# Will create a grid using the limits and creating 30 points between the limits example( 0,0   1,0    2,1  etc)
xx=np.linspace(xlim[0],xlim[1],30)
yy=np.linspace(ylim[0],ylim[1],30)
yy,xx=np.meshgrid(yy,xx) # Meshgrid will combine those 2 to form a grid
xy=np.vstack([xx.ravel(),yy.ravel()]).T # will create a stack of datapoints
z=clf.decision_function(xy).reshape(xx.shape) # 
ax.contour(xx,yy,z,colors='k',levels=[-1,0,1],alpha=0.5,linestyle=['--','-','--'])
ax.scatter(clf.support_vectors_[:,0],clf.support_vectors_[:,1],s=100,linewidth=1,facecolors='none')
plt.show()
```
