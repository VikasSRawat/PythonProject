```python
cur_x=5 # By default x = 3

rate=0.1 # The learning rate i.e., it will check for the local maxima and local minima by checking every 0.1 bars
# Note if we are dealing with increments of a 1000 and then it will have to check 1000/0.1 === 10000 bars
# But if we are dealing with increments of 0.01 then we might skip the maxima and minima entirely as 0.01/0.1 = 0.1 hence
# are skipping  over 0.01 bars

precision=0.5 # When to stop the algorithm when working with floating type of data i.e., max precision = 0.5

previous_step_size=1 # We will check the precision with this variable
max_iters=100000 #max iterations
iters=0# starting iterator

df=lambda x: 2*(x+5) # Gradeint of the function

while previous_step_size>precision and iters<max_iters:
    prev_x=cur_x# prev_x will have the current value
    cur_x=cur_x-rate*df(prev_x)# updating the current value with gradient descent algo
    prev_step_size=abs(cur_x-prev_x)# checking precision
    iters=iters+1
    print("Iterations = ",iters,"\nx value = ",cur_x)
    
print("Local minimum occurs at",cur_x)
```
